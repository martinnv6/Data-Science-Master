# Data Visualization

###Wine variety

```python
sns.countplot(wine['region_1'],order = wine['region_1'].value_counts().iloc[:10].index)
fig = plt.gcf()
fig.set_size_inches(20,20)
plt.title('Top 10 Region')
```

### Wine ratings

```python
sns.countplot(wine['points'])
sns.countplot(wine['points']).set_xticklabels(sns.countplot(wine['points']).get_xticklabels(), rotation=90, ha="right")
fig = plt.gcf()
fig.set_size_inches(13,13)
plt.title('Wine Rating')
```

### Relation between Price and Rating

```python
plt.figure(figsize=(10,8))
sns.lineplot(x='points',y='price',data=wine)
plt.title('Relation between Price and Rating')
plt.show()
```

### Pie-chart for taster name

```python
wine['taster_name'].value_counts().plot.pie(autopct='%1.1f%%',shadow=True,figsize=(10,10))
plt.show()
```

### Pie-chart for Country

```python
wine['country'].value_counts().plot.pie(autopct='%1.1f%%',shadow=True,figsize=(10,10))
plt.show()
```
